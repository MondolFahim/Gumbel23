```python
import numpy as np
import pandas as pd
from scipy.optimize import curve_fit
from sklearn.metrics import r2_score  # Import r2_score from scikit-learn
import matplotlib.pyplot as plt

# Load your data
file_path = 'C:\\Users\\USER\\Desktop\\conference\\bahadurbad.xlsx'
df = pd.read_excel(file_path)

# Assuming the peak flood data is in the 'Discharge(Cumec)' column
discharge = np.array(df['Discharge(Cumec)'])

# Assuming the years are in the 'Year' column
years = np.array(df['Year'])

# Define the exponential function
def exponential_function(x, a, b, c):
    return a * np.exp(b * x) + c

# Different initial guesses for parameters
initial_guesses = [1, -0.0001, 1]

# Fit the exponential model with adjusted initial guesses and increased maxfev
params, _ = curve_fit(exponential_function, years, discharge, p0=initial_guesses, maxfev=5000)

# Calculate predicted values
predicted_values_exp = exponential_function(years, *params)

# Print the R2 value for the exponential model
r2_exp = r2_score(discharge, predicted_values_exp)
print(f'R2 Value for Exponential Model: {r2_exp:.4f}')

# Plot observed data and the exponential model
plt.scatter(years, discharge, label='Observed Data')
plt.plot(years, predicted_values_exp, color='green', label=f'Exponential Model (R²={r2_exp:.4f})')
plt.xlabel('Year')
plt.ylabel('Discharge (Cumec)')
plt.legend()
plt.show()

# Print the optimal parameters
print(f'Optimal Parameters: a={params[0]}, b={params[1]}, c={params[2]}')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import genextreme, probplot, kstest

# Load data from Excel file
file_path = r'C:\Users\USER\Desktop\conference\bahadurbad.xlsx'
df = pd.read_excel(file_path)

# Assuming your Excel file has a column named 'Discharge(Cumec)' containing your data
observed_data = np.array(df['Discharge(Cumec)'])

# Fit Gumbel distribution
params = genextreme.fit(observed_data)
gumbel_distribution = genextreme(*params)

# Q-Q plot
probplot(observed_data, dist=gumbel_distribution, plot=plt)
plt.title('Q-Q Plot - Gumbel Distribution Fit')
plt.show()

# Kolmogorov-Smirnov test
ks_statistic, ks_p_value = kstest(observed_data, gumbel_distribution.cdf)
print(f"KS Statistic: {ks_statistic}")
print(f"P-Value: {ks_p_value}")
```
